# StackAPI

#### Import the necessary libraries here:

```python
import requests
import json
from stackapi import StackAPI
import datetime as dt
import time
```

#### Question 1: Find the questions and answers of last month.

```python
#StackAPI needs timestamp date format
#using module time.mktime to get timestamp format

print('01/01/20 in timestamp:',time.mktime(time.strptime('01/01/2020', '%d/%m/%Y')))
print('31/01/20 in timestamp:',time.mktime(time.strptime('31/01/2020', '%d/%m/%Y')))
```

```python
SITE = StackAPI('stackoverflow')
#importing January questions - timestamp not needed afterall 
#using dt.datetime fits API requirements
questions = SITE.fetch('questions', fromdate=dt.datetime(2020,1,1), todate=dt.datetime(2020,1,31))
questions
```

```python
#importing answers through API calls
answers = SITE.fetch('answers', fromdate=dt.datetime(2020,1,1), todate=dt.datetime(2020,1,31))
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
SITE = StackAPI('stackoverflow')

questions_score5 = SITE.fetch('questions', fromdate=1580166001, todate=1580252399,
                              min=5, tagged='python', sort='votes')

print(json.dumps(questions_score5, indent=4))
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
SITE = StackAPI('stackoverflow')

answers_id = SITE.fetch('answers', ids=[6784 , 6473])
print(json.dumps(answers_id, indent=4))
```
