## Challenge 2: Count Commits
Count how many commits were made in the month of october of 2018.
#### 1. Obtain all the commits made in October 2018 via API, which is a JSON array that contains multiple commit objects.

`````python
url_iron = f'https://api.github.com/repos/{iron_repo}/commits'

params = {'since':'2018-10-01T00:00:00Z', 'until':'2018-10-31T23:59:00Z'}

response = requests.get(url_iron, auth=(user, token), params=params)

#printing requests answer code
print(response)

result = response.json()

print(json.dumps(result, indent=8))
`````

#### 2. Count how many commit objects are contained in the array.

`````python
list_of_commits = []

for i in result:
    list_of_commits.append(i['commit'])

print('there has been',len(list_of_commits),'commits in October 2018')
`````
